## B.4 Apply functions to pandas

### B.4.1 Basics

```python
def basics(data):
    print('Head')
    print(data.head(2))
    print('')
    print('Tail')
    print(data.tail(2))
    print('')
    print('Columns')
    print(data.columns[:3])
    print('')
    print('Info')
    print(data.info())

basics(df)
```

### B.4.2 subset

```python
def subset(countries, columns):
    out = df.query(f"country in {countries}")[columns]
    return out
subset(
    countries = ['Japan', 'Mexico'],
    columns = ['country', 'date']
)
```
